## 2. Simulación: Índices LEPI aplicados al aula

```
# ─── Simulación de un grupo de 35 estudiantes ────────
np.random.seed(42)
n_est = 35

df_grupo = pd.DataFrame({
    'estudiante':     [f'EST-{str(i).zfill(2)}' for i in range(1, n_est+1)],
    'semana':         np.tile(np.arange(1, 8), 5)[:n_est],
    # Dimensiones LEPI del estudiante (escala 1-5)
    'lid_participacion':  np.round(np.random.normal(3.6, 0.7, n_est).clip(1, 5), 1),
    'prax_reflexion':     np.round(np.random.normal(3.4, 0.8, n_est).clip(1, 5), 1),
    'inn_creatividad':    np.round(np.random.normal(3.5, 0.7, n_est).clip(1, 5), 1),
    'asistencia_pct':     np.round(np.random.uniform(60, 100, n_est), 1),
    'nota_pretest':       np.round(np.random.normal(3.2, 0.7, n_est).clip(1, 5), 1),
    'nota_postest':       np.round(np.random.normal(4.1, 0.5, n_est).clip(1, 5), 1),
})

df_grupo['ILEPI_est'] = (
    df_grupo['lid_participacion'] * 0.35 +
    df_grupo['prax_reflexion']    * 0.35 +
    df_grupo['inn_creatividad']   * 0.30
).round(2)

df_grupo['ganancia'] = (df_grupo['nota_postest'] - df_grupo['nota_pretest']).round(2)
df_grupo['alerta']   = df_grupo['asistencia_pct'].apply(
    lambda x: '🔴 Riesgo' if x < 75 else ('🟡 Atención' if x < 85 else '🟢 OK')
)

print("Indicadores del grupo — Metodología LEPI:")
print(df_grupo[['estudiante','ILEPI_est','asistencia_pct',
                'nota_pretest','nota_postest','ganancia','alerta']].head(10))
print(f"\nGanancia promedio de aprendizaje: +{df_grupo['ganancia'].mean():.2f} puntos")
print(f"ILEPI promedio del grupo: {df_grupo['ILEPI_est'].mean():.2f}/5.0")
```

```
fig, axes = plt.subplots(2, 2, figsize=(14, 10))
fig.suptitle('Dashboard Docente — Grupo LEPI\nCristhian Díaz Esteban',
             fontsize=13, fontweight='bold')

# Pre vs Post test
axes[0,0].hist(df_grupo['nota_pretest'],  bins=12, alpha=0.7, color='#E74C3C', label='Pre-test')
axes[0,0].hist(df_grupo['nota_postest'],  bins=12, alpha=0.7, color='#27AE60', label='Post-test')
axes[0,0].axvline(df_grupo['nota_pretest'].mean(),  color='red',   linestyle='--', alpha=0.8)
axes[0,0].axvline(df_grupo['nota_postest'].mean(),  color='green', linestyle='--', alpha=0.8)
axes[0,0].set_title(f"Pre-test vs Post-test\n(Ganancia: +{df_grupo['ganancia'].mean():.2f} pts)")
axes[0,0].legend()

# Alertas de asistencia
alertas = df_grupo['alerta'].value_counts()
col_al = {'🟢 OK':'#27AE60','🟡 Atención':'#F39C12','🔴 Riesgo':'#E74C3C'}
axes[0,1].bar(alertas.index, alertas.values,
              color=[col_al.get(k,'gray') for k in alertas.index], edgecolor='white')
axes[0,1].set_title('Alertas de Asistencia\n(momento VER del LEPI)')

# ILEPI vs Ganancia
sc = axes[1,0].scatter(df_grupo['ILEPI_est'], df_grupo['ganancia'],
                        c=df_grupo['asistencia_pct'], cmap='RdYlGn',
                        s=80, alpha=0.8)
plt.colorbar(sc, ax=axes[1,0], label='Asistencia %')
z = np.polyfit(df_grupo['ILEPI_est'], df_grupo['ganancia'], 1)
x_l = np.linspace(df_grupo['ILEPI_est'].min(), df_grupo['ILEPI_est'].max(), 50)
axes[1,0].plot(x_l, np.poly1d(z)(x_l), 'r--', alpha=0.7)
r, p = stats.pearsonr(df_grupo['ILEPI_est'], df_grupo['ganancia'])
axes[1,0].set_xlabel('ILEPI Estudiante'); axes[1,0].set_ylabel('Ganancia de Aprendizaje')
axes[1,0].set_title(f'ILEPI → Ganancia\nr={r:.3f} p={p:.3f}')

# Distribución ILEPI
axes[1,1].hist(df_grupo['ILEPI_est'], bins=15, color='#2E75B6', edgecolor='white', alpha=0.85)
axes[1,1].axvline(df_grupo['ILEPI_est'].mean(), color='red', linestyle='--',
                   label=f"Media: {df_grupo['ILEPI_est'].mean():.2f}")
axes[1,1].axvspan(1, 3.0, alpha=0.08, color='red')
axes[1,1].axvspan(3.0, 3.5, alpha=0.08, color='orange')
axes[1,1].axvspan(3.5, 5.0, alpha=0.08, color='green')
axes[1,1].set_title('Distribución ILEPI del Grupo')
axes[1,1].legend()

plt.tight_layout()
plt.show()
```
