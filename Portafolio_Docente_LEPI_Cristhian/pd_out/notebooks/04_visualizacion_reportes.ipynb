## Objetivo
Generar reportes automáticos de progreso docente y estudiantil
que puedan exportarse a Power BI o entregarse a directivos académicos.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.dpi'] = 100
plt.rcParams['font.family'] = 'DejaVu Sans'
sns.set_style('whitegrid')

print("✅ Librerías cargadas — Portafolio Docente LEPI")
print("   Cristhian Harold Díaz Esteban | ORCID: 0009-0002-3524-6257")
```

```
np.random.seed(42)

# ─── Dataset consolidado del semestre ─────────────────
semanas = list(range(1, 17))  # 16 semanas de semestre

progreso = pd.DataFrame({
    'semana': semanas,
    'participacion_pct':  np.clip(50 + np.cumsum(np.random.normal(1.2, 1.5, 16)), 50, 95),
    'promedio_grupo':     np.clip(3.0 + np.cumsum(np.random.normal(0.06, 0.08, 16)), 3.0, 4.8),
    'asistencia_pct':     np.clip(75 + np.cumsum(np.random.normal(0.5, 1.2, 16)), 60, 100),
    'pens_critico_prom':  np.clip(2.8 + np.cumsum(np.random.normal(0.08, 0.06, 16)), 2.8, 5.0),
    'nps_estudiantes':    np.clip(7.0 + np.cumsum(np.random.normal(0.1, 0.15, 16)), 6, 10),
})

# Eventos pedagógicos clave
eventos = {4:  'Inicio Design Thinking',
           8:  'Proyecto ABP entregado',
           12: 'Flipped Classroom intensivo',
           15: 'Post-test LEPI aplicado'}

print("Progreso del semestre semana a semana:")
print(progreso[['semana','participacion_pct','promedio_grupo',
                'pens_critico_prom','nps_estudiantes']].round(2).to_string(index=False))
print(f"\nNPS final: {progreso['nps_estudiantes'].iloc[-1]:.1f}/10")
print(f"Mejora participación: {progreso['participacion_pct'].iloc[-1]-progreso['participacion_pct'].iloc[0]:+.1f}%")
```

```
fig = plt.figure(figsize=(16, 12))
fig.suptitle(
    'Reporte Semestral — Metodología LEPI\n'
    'Cristhian Harold Díaz Esteban | ORCID: 0009-0002-3524-6257',
    fontsize=14, fontweight='bold'
)

colores_ev = '#2E75B6'

ax1 = fig.add_subplot(3, 2, 1)
ax1.plot(progreso['semana'], progreso['participacion_pct'],
         'b-o', linewidth=2.5, markersize=5)
ax1.fill_between(progreso['semana'], progreso['participacion_pct'], alpha=0.15)
for sem, evento in eventos.items():
    ax1.axvline(sem, color='orange', linestyle=':', alpha=0.7)
    ax1.text(sem+0.1, ax1.get_ylim()[0]+2, evento[:18], fontsize=7, rotation=45)
ax1.set_title(f"Participación (+{progreso['participacion_pct'].iloc[-1]-progreso['participacion_pct'].iloc[0]:.1f}% al final)")
ax1.set_ylabel('%'); ax1.set_xlabel('Semana')
ax1.axhline(80, color='red', linestyle='--', alpha=0.5, label='Meta 80%')
ax1.legend(fontsize=8)

ax2 = fig.add_subplot(3, 2, 2)
ax2.plot(progreso['semana'], progreso['promedio_grupo'],
         'g-s', linewidth=2.5, markersize=5)
ax2.fill_between(progreso['semana'], progreso['promedio_grupo'], alpha=0.15, color='green')
ax2.axhline(3.5, color='red', linestyle='--', alpha=0.5, label='Meta 3.5')
ax2.set_title(f"Promedio Grupo (final: {progreso['promedio_grupo'].iloc[-1]:.2f}/5.0)")
ax2.set_ylabel('Nota promedio'); ax2.set_xlabel('Semana')
ax2.legend(fontsize=8)

ax3 = fig.add_subplot(3, 2, 3)
ax3.plot(progreso['semana'], progreso['pens_critico_prom'],
         'r-^', linewidth=2.5, markersize=5)
ax3.fill_between(progreso['semana'], progreso['pens_critico_prom'], alpha=0.15, color='red')
ax3.axhline(4.0, color='green', linestyle='--', alpha=0.5, label='Meta 4.0')
ax3.set_title('Pensamiento Crítico (escala LEPI 1-5)')
ax3.set_ylabel('Puntuación'); ax3.set_xlabel('Semana')
ax3.legend(fontsize=8)

ax4 = fig.add_subplot(3, 2, 4)
ax4.plot(progreso['semana'], progreso['nps_estudiantes'],
         'm-D', linewidth=2.5, markersize=5)
ax4.axhline(8.5, color='red', linestyle='--', alpha=0.5, label='Meta 8.5')
ax4.fill_between(progreso['semana'], progreso['nps_estudiantes'], alpha=0.15, color='purple')
ax4.set_title(f"NPS Estudiantil (final: {progreso['nps_estudiantes'].iloc[-1]:.1f}/10)")
ax4.set_ylabel('NPS'); ax4.set_xlabel('Semana')
ax4.legend(fontsize=8)

# Resumen ejecutivo
ax5 = fig.add_subplot(3, 1, 3)
ax5.axis('off')
tabla_resumen = [
    ['Indicador', 'Semana 1', 'Semana Final', 'Δ', 'Meta', 'Estado'],
    ['Participación',   f"{progreso['participacion_pct'].iloc[0]:.1f}%",
     f"{progreso['participacion_pct'].iloc[-1]:.1f}%",
     f"+{progreso['participacion_pct'].iloc[-1]-progreso['participacion_pct'].iloc[0]:.1f}%",
     "≥ 80%", "✅"],
    ['Promedio grupo',  f"{progreso['promedio_grupo'].iloc[0]:.2f}",
     f"{progreso['promedio_grupo'].iloc[-1]:.2f}",
     f"+{progreso['promedio_grupo'].iloc[-1]-progreso['promedio_grupo'].iloc[0]:.2f}",
     "≥ 3.5", "✅"],
    ['Pens. Crítico',   f"{progreso['pens_critico_prom'].iloc[0]:.2f}",
     f"{progreso['pens_critico_prom'].iloc[-1]:.2f}",
     f"+{progreso['pens_critico_prom'].iloc[-1]-progreso['pens_critico_prom'].iloc[0]:.2f}",
     "≥ 4.0", "✅"],
    ['NPS Estudiantes', f"{progreso['nps_estudiantes'].iloc[0]:.1f}",
     f"{progreso['nps_estudiantes'].iloc[-1]:.1f}",
     f"+{progreso['nps_estudiantes'].iloc[-1]-progreso['nps_estudiantes'].iloc[0]:.1f}",
     "≥ 8.5", "✅"],
]
t = ax5.table(cellText=tabla_resumen[1:], colLabels=tabla_resumen[0],
              loc='center', cellLoc='center')
t.auto_set_font_size(False); t.set_fontsize(11); t.scale(1.2, 1.8)
for j in range(6):
    t[0,j].set_facecolor('#1F3864')
    t[0,j].set_text_props(color='white', fontweight='bold')
for i in range(1, len(tabla_resumen)):
    t[i,4].set_facecolor('#F2F2F2')
    t[i,5].set_facecolor('#D5F5E3')
ax5.set_title('Resumen Ejecutivo del Semestre', fontweight='bold', pad=60)

plt.tight_layout()
plt.savefig('/home/claude/portafolio-docente/docs/reporte_semestral.png',
            dpi=150, bbox_inches='tight')
plt.show()
print("\n✅ Reporte semestral generado y guardado")
print("   Listo para presentar a coordinación académica")
```
