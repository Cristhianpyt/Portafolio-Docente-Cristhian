## Objetivo
Evaluar el impacto de la metodología LEPI mediante análisis estadístico
antes/después, replicando la metodología de mi investigación publicada.


```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import seaborn as sns
from scipy import stats
import warnings
warnings.filterwarnings('ignore')

plt.rcParams['figure.dpi'] = 100
plt.rcParams['font.family'] = 'DejaVu Sans'
sns.set_style('whitegrid')

print("✅ Librerías cargadas — Portafolio Docente LEPI")
print("   Cristhian Harold Díaz Esteban | ORCID: 0009-0002-3524-6257")
```

```
# ─── Datos reales de la investigación + simulación extendida ───
# Instituciones donde se ha implementado el LEPI

instituciones = pd.DataFrame({
    'institucion':    ['UNIMINUTO Bquilla','UNIMINUTO BGA','UNIMINUTO Bogotá',
                       'Univ. Compensar','UIS Mentoring','Javeriana MEN',
                       'Corpoeducación','Colegio Atenas'],
    'tipo':           ['Universidad','Universidad','Universidad',
                       'Universidad','Universidad','Universidad',
                       'ONG','Colegio'],
    'participantes':  [80, 65, 120, 200, 85, 100, 150, 45],
    'antes_participacion': [52, 55, 48, 60, 58, 62, 45, 50],
    'despues_participacion':[67, 72, 65, 72, 81, 81, 61, 68],
    'antes_pensamiento_crit':[55, 50, 48, 58, 55, 60, 42, 48],
    'despues_pensamiento_crit':[75, 72, 69, 78, 77, 81, 65, 70],
})

instituciones['delta_participacion'] = (
    instituciones['despues_participacion'] -
    instituciones['antes_participacion']
)
instituciones['delta_pensamiento'] = (
    instituciones['despues_pensamiento_crit'] -
    instituciones['antes_pensamiento_crit']
)

print("Impacto del LEPI por institución:")
print(instituciones[['institucion','participantes',
                      'delta_participacion','delta_pensamiento']].to_string(index=False))
print(f"\nDelta participación promedio: +{instituciones['delta_participacion'].mean():.1f}%")
print(f"Delta pensamiento crítico:     +{instituciones['delta_pensamiento'].mean():.1f}%")
```

```
# ─── Prueba estadística (t-test pareado) ─────────────
from scipy.stats import ttest_rel

indicadores = [
    ('Participación',  'antes_participacion',     'despues_participacion'),
    ('Pens. Crítico',  'antes_pensamiento_crit',  'despues_pensamiento_crit'),
]

print("Análisis estadístico — Prueba t pareada:")
print("─"*65)
print(f"{'Indicador':<20} {'Antes':>8} {'Después':>10} {'Δ':>8} {'p-valor':>10} {'Sig.'}")
print("─"*65)

for nombre, ca, cd in indicadores:
    a, d   = instituciones[ca].values, instituciones[cd].values
    t, p   = ttest_rel(d, a)
    delta  = d.mean() - a.mean()
    pooled = np.sqrt((a.std()**2 + d.std()**2)/2)
    cohd   = delta / pooled if pooled > 0 else 0
    sig    = '✅***' if p<0.001 else '✅**' if p<0.01 else '✅*' if p<0.05 else '⚠️n.s.'
    print(f"{nombre:<20} {a.mean():>8.1f}% {d.mean():>9.1f}% {delta:>+7.1f}% {p:>9.4f} {sig}")
print("─"*65)
print("*** p<0.001 | ** p<0.01 | * p<0.05")

# Visualización del impacto
fig, axes = plt.subplots(1, 2, figsize=(14, 6))
fig.suptitle('Impacto LEPI — Análisis Antes/Después\nCristhian Díaz Esteban (2024)',
             fontsize=13, fontweight='bold')

for i, (nombre, ca, cd) in enumerate(indicadores):
    antes   = instituciones[ca].values
    despues = instituciones[cd].values
    x       = np.arange(len(instituciones))
    width   = 0.35
    axes[i].bar(x-width/2, antes,   width, color='#E74C3C', alpha=0.8, label='Antes')
    axes[i].bar(x+width/2, despues, width, color='#27AE60', alpha=0.8, label='Después')
    axes[i].set_xticks(x)
    axes[i].set_xticklabels(
        [n[:12] for n in instituciones['institucion']], rotation=35, ha='right', fontsize=8
    )
    axes[i].set_title(f'{nombre} (%)\nΔ={despues.mean()-antes.mean():+.1f}%')
    axes[i].set_ylabel('%'); axes[i].legend()
    axes[i].set_ylim(0, 100)

plt.tight_layout()
plt.show()
```
